```python
import unicodedata # 

def remove_accents(input_str): # to remove accents  eg: nin~()a to nina
    
    nfkd_form = unicodedata.normalize('NFKD', input_str) #nfkd_form :cariable
    
    return u"".join([c for c in nfkd_form if not unicodedata.combining(c)]) #u: variable (unique code)
```

```python
characters = ["Shiva","Sati","Bhadra","Veerbhadra","Nandi","Krittika", "Parvateshwar", "Anandmayi",
              "Bhrahaspati","Daksha","Kanakhala","Bhagirath","Veerini", "Ayurvati","Hooded Naga","Dilipa" ] 
```

```python
with io.open('../Data/meluha.txt', encoding='utf8') as f_in:
    Meluha_text = f_in.read()
    
Meluha = remove_accents(Meluha_text)
```

```python
Meluha_chapters = Meluha.split('CHAPTER')[1:]
```

```python
len(Meluha_chapters)
Meluha_chapters = Meluha_chapters[1:]
```

```python
Meluha_chapters = Meluha_chapters[1:]
```

```python
for CHAPTER in Meluha_chapters:  # try to include name of the chapter alongside the length
    print(len(CHAPTER))
```

```python
chapters = ["He has come!","Land of Pure Life","She Enters His Life","Abode of the Gods",
"Tribe of Brahma","Vikarma, the Carriers of Bad Fate","Lord Ram’s Unfinished Task",
"Drink of the Gods","The Hooded Figure Returns","Neelkanth Unveiled","Journey through Meluha",
"Blessings of the Impure","Pandit of Mohan Jo Daro","Trial by Fire","The Sun & Earth" 
"The Battle of Koonj","Sati and the Fire Arrow","Love Realised","Attack on Mandar",
"Preparation for War","Empire of Evil","Dharmayudh, the Holy War","A Stunning Revelation",
"Island of the Individual","The Question of Questions" ] #list of Chapters

counts = ["Chapters" + '\t'+'\t'.join(characters)] # 
#print(counts)
for i in range(len(Meluha_chapters)):
    line = CHAPTER[i]
    for ch in characters: #21 characters in Ramayana we have taken
        line = line + '\t' + str(Meluha_chapters[i].count(ch))
    counts.append(line)
            
with open("../Data/char_count.xls",'w') as f_out:  # writing a character count in the file named char_count.txt
    for line in counts:
        f_out.write(line+'\n') # list which is written out line by line to the file
    
```

```python
df = pd.read_csv("../Data/char_count.xls", sep = '\t') #df: variabel name ;read the above saved file as dataframe
# here csv file is seperated by tabs so   sep = '\t' used
#print(df) # the column has... bcoz its long;feature of jupiter
df
```
